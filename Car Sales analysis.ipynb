# CHART11

```python
import pandas as pd

# Count the number of sales for each model by company
sales_counts = cs_df1.groupby(['Company', 'Model']).size().reset_index(name='Sales')

# Count the number of unique models for each company
model_counts = sales_counts.groupby('Company')['Model'].nunique().reset_index()

# Filter companies with more than 5 models
companies_with_more_than_5 = model_counts[model_counts['Model'] > 5]['Company']

# Filter the original sales data to only include those companies
filtered_sales = sales_counts[sales_counts['Company'].isin(companies_with_more_than_5)]

# Prepare a dictionary to store the least-selling models for each company
least_selling_models_dict = {company: [] for company in companies_with_more_than_5}

# Get the 2 least-selling models for each company
for company in companies_with_more_than_5:
    company_models = filtered_sales[filtered_sales['Company'] == company]
    least_models = company_models.nsmallest(2, 'Sales')
    
    # Add body style information
    for model in least_models['Model']:
        body_style = cs_df1.loc[cs_df1['Model'] == model, 'Body Style'].iloc[0]
        least_selling_models_dict[company].append(f"{model} ({body_style})")

# Create a DataFrame from the dictionary
least_selling_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in least_selling_models_dict.items()]))

# Rename index to 1 and 2
least_selling_df.index = [1, 2]

# Style the DataFrame to add lines, bold column names, and adjust column width
styled_table = least_selling_df.style.set_table_attributes('style="border-collapse: collapse; width: 100%;"') \
    .set_properties(**{'border': '1px solid black'}) \
    .set_table_styles([
        {'selector': 'th', 'props': [('font-weight', 'bold'), ('border', '1px solid black'), ('background-color', 'grey')]}
    ]) \
    .set_properties(subset=[least_selling_df.columns[0]], **{'min-width': '50px'})  # Adjust first column width

# Display the styled table
display(styled_table)
```

**Insight** 

-Helps us Understand which are the 2 least perfoming cars in the market brand wise

# CHART12

```python
import pandas as pd

# Count the number of sales for each model by company
model_sales_counts = cs_df1.groupby(['Company', 'Model']).size().reset_index(name='Sales')

# Assuming the body style information is in the DataFrame
# We need to add the body style to the model
body_style_info = cs_df1[['Model', 'Body Style']].drop_duplicates()
model_sales_counts = model_sales_counts.merge(body_style_info, on='Model')

# Get the top 2 most sold models for each company
top_models = model_sales_counts.loc[model_sales_counts.groupby('Company')['Sales'].nlargest(2).reset_index()['level_1']]

# Create a new column for Model and Body Style
top_models['Model with Style'] = top_models['Model'] + ' (' + top_models['Body Style'] + ')'

# Prepare a dictionary to store the top models for each company
top_models_dict = {}

# Populate the dictionary
for company in top_models['Company'].unique():
    top_models_dict[company] = top_models[top_models['Company'] == company]['Model with Style'].tolist()

# Create a DataFrame from the dictionary
top_models_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in top_models_dict.items()]))

# Rename index to 1 and 2
top_models_df.index = [1, 2]

# Display 5 brands per table
num_brands_per_table = 5
num_tables = (len(top_models_df.columns) + num_brands_per_table - 1) // num_brands_per_table

for i in range(num_tables):
    # Slice the DataFrame for the current table
    sliced_df = top_models_df.iloc[:, i * num_brands_per_table:(i + 1) * num_brands_per_table]
    
    # Style the sliced DataFrame to add lines and bold column names
    styled_table = sliced_df.style.set_table_attributes('style="border-collapse: collapse; width: 100%;"') \
        .set_properties(**{'border': '1px solid black'}) \
        .set_table_styles([
            {'selector': 'th', 'props': [('font-weight', 'bold'), ('border', '1px solid black'), ('background-color', 'grey')]}  # Header style
        ]) \
        .set_properties(subset=[sliced_df.columns[0]], **{'min-width': '50px'})  # Adjust the first column width

    # Display the styled table
    display(styled_table)
```

**Insight** 

-Helps us Understand which are the 2 Most perfoming cars and which body type in the market brand wise

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a new column that combines Model and Body Style
cs_df1['Model_BodyStyle'] = cs_df1['Model'] + ' (' + cs_df1['Body Style'] + ')'

# Count the number of sales for each Model_BodyStyle based on Transmission type
transmission_counts = cs_df1.groupby(['Model_BodyStyle', 'Transmission']).size().unstack(fill_value=0)

# Create side-by-side horizontal bar charts
plt.figure(figsize=(20, 50))  # Further increase the height for better spacing

# Horizontal bar chart for Automatic Transmission
plt.subplot(1, 2, 1)
automatic_sales = transmission_counts['Auto'].sort_values(ascending=False)
automatic_sales.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Sales by Model (Automatic Transmission)')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Model (Body Style)')
plt.grid(axis='x')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest sales at the top

# Add x-axis labels on top
for index, value in enumerate(automatic_sales):
    plt.text(value, index, str(value), va='center', ha='left', fontsize=10)

# Horizontal bar chart for Manual Transmission
plt.subplot(1, 2, 2)
manual_sales = transmission_counts['Manual'].sort_values(ascending=False)
manual_sales.plot(kind='barh', color='orange', edgecolor='black')
plt.title('Sales by Model (Manual Transmission)')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Model (Body Style)')
plt.grid(axis='x')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest sales at the top

# Add x-axis labels on top
for index, value in enumerate(manual_sales):
    plt.text(value, index, str(value), va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set(style="whitegrid")

# Function to create a bar plot for the relationship between two categorical columns
def plot_relationship(column1, column2):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=cs_df1, x=column1, hue=column2, palette='pastel')
    plt.title(f'Relationship between {column1} and {column2}')
    plt.xlabel(column1)
    plt.ylabel('Count')
    
    # Adjust the legend position
    plt.legend(title=column2, loc='upper right', bbox_to_anchor=(1.15, 1))
    
    plt.xticks(rotation=45)  # Rotate x labels for better readability
    plt.show()

# Example: Plotting the relationship between 'Transmission' and 'Body Style'
plot_relationship('Transmission', 'Body Style')
```

# Recommendation

**-Focus on SUV and Hatchback Offerings:**

*Given that SUVs (26.66%) and hatchbacks (25.63%) are the most popular segments, dealerships should prioritize stocking and promoting these models. Highlighting features that cater to consumer preferences can drive sales further.**

**-Capitalize on Growing Demand for Automatic Transmission:**

*As both male and female customers are increasingly opting for automatic transmission (52.52% for males, 52.84% for females), dealerships should expand their offering of automatic transmission models, especially in SUVs and Hatchbacks. Marketing campaigns that emphasize the ease and convenience of automatic vehicles could drive more sales.**

**-Shift Away from Hardtops:**

*Since hardtops are the least preferred across regions and transmission types, dealerships should limit their investment in hardtop models. Instead, focus on models with higher demand like SUVs and Hatchbacks.**

**-Expand Sales of Best-Selling Models:**

*The Volkswagen Passat, Prizm, Silhouette, and LS400 are among the top-selling car models. Dealerships should make sure to stock more of these high-demand models and market them as best-sellers to potential customers. Highlighting their popularity could lead to faster sales.*

**-Target the Austin Market:**

*Austin is the largest market, with over 4000 units sold. Dealerships in or near this region should focus their marketing and sales efforts there, possibly introducing special regional offers or hosting events to capitalize on this significant customer base.*

**-Promote Hatchbacks with Auto Transmission and SUVs with Manual:**

*Based on customer preferences, dealerships should highlight that Hatchbacks are more popular in auto transmission and SUVs are preferred with manual transmission. Tailoring marketing messages and promotions around these preferences can help attract more targeted buyers.*

**-Increase Diesel Car Inventory:**

*Diesel cars make up 52.59% of sales due to their long life, reliability, and durability. Dealerships should prioritize diesel models in their inventory, especially for customers looking for long-term vehicles, while still maintaining a good stock of petrol cars for those with different preferences.*

**-Optimize Inventory for the Mid-Range Market:**

*With the average price of cars sold at approximately $28,090, dealerships should focus on stocking vehicles within or slightly above this price range. Offering attractive financing options or special deals on cars near this price point could appeal to a larger portion of customers and drive more sales.*

**-Stock Up on Popular Shades:**

*Since 'pale white' is the most preferred color followed by 'Black' among customers, dealerships should prioritize increasing inventory for cars in this color. Promoting pale white models as trendy or premium can help attract even more buyers. Dealerships should also highlight this preference in their marketing campaigns to meet customer demand effectively.**
